```python
import numpy as np
import warnings 
warnings.filterwarnings("ignore")

import fda.bezier as bz
```

### Calculate and Plot Bezier Curve

**Example 1:**

```python
# Set control points
control_points = np.array([
    [10,2],
    [7,5],
    [8, 9]
])


# Print out Bezier coordinates for the given control points
bezier_obj = bz.BezierCurve(control_points)
bezier_coordinates = bezier_obj.get_bezier_curve()
for ind, coord in enumerate(bezier_coordinates):
    print(f"{ind+1}: {coord}")
```

```python
# Plot component graphs
bezier_obj.plot_component_curve()
```

```python
# Plot Bezier curve
bezier_obj.plot_bezier_curve()
```
